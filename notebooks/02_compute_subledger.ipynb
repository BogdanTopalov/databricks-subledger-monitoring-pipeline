### Create the daily balances

```python
spark.sql("USE demo_subledger_monitoring")

# Recompute subledger_daily from transactions_raw
spark.sql("TRUNCATE TABLE subledger_daily")

spark.sql("""
    INSERT INTO subledger_daily
    SELECT
        posting_date,
        account_id,
        currency,
        CAST(SUM(amount) AS DECIMAL(18,2)) AS total_amount,
        COUNT(*) AS transaction_count,
        current_timestamp() AS computed_ts
    FROM transactions_raw
    GROUP BY posting_date, account_id, currency
""")

display(spark.sql("""
    SELECT *
    FROM subledger_daily
    ORDER BY posting_date, account_id, currency
"""))
```
